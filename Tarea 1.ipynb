# Capítulo 1. Trabajando con números.

Todos los ejercicios de esta tarea se pueden realizar consultando el capítulo 1 del libro. En los ejercicios se trabaja con números enteros y flotantes. También se da una introducción al trabajo con fracciones, para lo que es necesario cargar la biblioteca `fractions`. La parte de números complejos del libro se puede obviar pues no hay ningún ejercicio que los necesite.


Aunque se proponen 11 ejercicios solamente se deben realizar 10 de ellos. El último está puesto para que si te atascas en un determinado ejercicio, tengas al oportunidad de realizar otro.

### 1.- Operaciones básicas (y Python puede equivocarse)

Realiza las siguientes operaciones. Señala si alguna de las operaciones no es matemáticamente correcta. Escribir la conclusión como un comentario en la celda donde esté la operación en la que Python no es matemáticamente correcto.

$$
a)\,34 + 67 \qquad b)\, 56 \cdot (-45) \cdot 7.1 \qquad c)\, 0.1 + 0.2
$$

### 2.- Cociente y resto.

Calcula el cociente y el resto de la división.

$$
344643 / 346
$$

### 3.- Potencias

Calcular las potencias.

$$
a)\, 2^5 \qquad b)\, (-3)^5 \qquad c)\, 2^{333}
$$

### 4.- Calculando radicales empleando potencias.

Recordando que las raíces en realidad son potencias, calcula las siguientes raíces.

$$
a)\, \sqrt{78} \qquad b)\, \sqrt[4]{2^7}
$$

### 5.- Tipos de datos numéricos en Python.

Encuentra el tipo de dato que asocia Python a los siguientes números.

$$
a)\, 8945 \qquad b)\, 2.0 \qquad c)\, 5.78
$$

### 6.- Redondeo y truncamiento.

Redondea y trunca los siguientes números (para redondear se puede emplear la función `round()` y para truncar la función `int()`). Explica la diferencia entre ambas funciones, como un comentario, en la última celda del ejercicio.

$$
a)\, 4.8 \qquad b)\, 2.1
$$

### 7.- El módulo `fractions`.

Utilizando el módulo `fractions` realiza la siguiente operación con fracciones. Si no se utiliza la función `print()` el resultado se presenta en la misma forma en la que se escriben las fracciones en Python. Escribe la fracción resultante dentro de la función `print()` y comprueba como mejora la presentación.

$$
\frac{3}{4} + \frac{6}{5} \cdot \frac{2}{7}
$$

### 8.- Programa que calcula cuadrados.

Crea un programa (o función) que pida un número entero al usuario, que convierta la entrada de `string` a `int` y que finalmente calcule el cuadrado del número en cuestión. Recuerda que la función `input()` siempre devuelve una cadena.

### 9.- Conversión de pulgadas a centímetros.

Crear un programa (o función) que convierta de pulgadas a centímetros. Debe pedir al usuario el número de pulgadas y devolver el número de centímetros (recuerda que una pulgada equivale a 2.54 cm)

### 10.- Una ecuación de segundo grado incompleta.

Crea un programa (o función) que resuelva la ecuación de segundo grado incompleta de la forma $ax^2 + bx$. El programa debe de pedir el valor de la $a$ y de la $b$ y devolver las dos soluciones. Ten en cuenta que tanto la $a$ como la $b$ pueden ser números decimales.

### 11.- Un problema con bucles.

Crea un programa que imprima el cuadrado de todos los números (enteros) entre el 1 y el 100.
